K-Means Clustering
FOR HEART DISEASE DATASET

```python
import numpy as np 
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
heart=pd.read_csv('heart.csv')
heart.head(5)
```

```python
heart.describe(include='all')
```

```python
dist = []

for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 15, random_state = 0)
    kmeans.fit(heart)
    dist.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 15), dist)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('DIST') 
plt.show()
```

```python
X=heart.iloc[:,:].values
```

```python
X.shape
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0,4], s = 100, c = 'purple', label = '0')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 4], s = 100, c = 'blue', label = '1')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 4], s = 100, c = 'orange', label = '2')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 4], s = 100, c = 'red', label = '3')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 4], s = 100, c = 'green', label = '4')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,4], s = 100, c = 'black', label = 'Centroids')

plt.title('Chances of getting Heart Disease')
plt.xlabel('Age ')
plt.ylabel('Cholestrol')
plt.legend()
plt.show()
```

FOR IRIS DATASET

```python
iris =pd.read_csv('iris.csv')
iris.head(5)
```

```python
iris.describe(include='all')
```

```python
iris.shape
```

```python
Y = iris.iloc[:, [0,1,2,3]].values
```

```python
dist1 = []

for i in range(1, 15):
    kmeans1 = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 15, random_state = 0)
    kmeans1.fit(Y)
    dist1.append(kmeans1.inertia_)
```

```python
plt.plot(range(1, 15), dist1)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('DIST') 
plt.show()
```

```python

kmeans1 = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans1 = kmeans1.fit_predict(Y)
iris['species']=y_kmeans1
```

```python
plt.scatter(Y[y_kmeans1 == 0, 0], Y[y_kmeans1 == 0,1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(Y[y_kmeans1 == 1, 0], Y[y_kmeans1 == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolor')
plt.scatter(Y[y_kmeans1 == 2, 0], Y[y_kmeans1 == 2, 1], s = 100, c = 'orange', label = 'Iris-virginica')


plt.scatter(kmeans1.cluster_centers_[:, 0], kmeans1.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroids')


plt.title('K-MEANS CLUSTERING FOR IRIS DATASET')
plt.legend()
plt.show()
```

FOR SYNTHETIC DATASET

```python
from sklearn.datasets import make_blobs
X3,y3 =make_blobs(n_samples=500,n_features=2,centers=3,random_state=42)
```

```python
dist2 = []

for i in range(1, 15):
    kmeans3 = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 15, random_state = 0)
    kmeans3.fit(X3)
    dist2.append(kmeans3.inertia_)
```

```python
plt.plot(range(1, 15), dist2)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('DIST') 
plt.show()
```

```python
kmeans3 = KMeans(n_clusters=3)
kmeans3.fit(X3)
```

```python
df = pd.DataFrame(data=X3, columns=['x1', 'x2'])
df['predicted_cluster'] = kmeans3.labels_
```

```python
print("The predicted column name is:", 'predicted_cluster')
```

```python
kmeans3 = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans3 = kmeans3.fit_predict(X3)
df['predicted_cluster']=y_kmeans3
```

```python
plt.scatter(X3[:, 0], X3[:, 1], c=kmeans3.labels_)
plt.scatter(kmeans3.cluster_centers_[:, 0], kmeans3.cluster_centers_[:, 1], s=50, color='green')

plt.title('K-MEANS OF SYNTEHTIC DATASET')
plt.show()
```

DBSCAN CLUSTERING


FOR HEART DISEASE DATASET


```python
from sklearn.cluster import DBSCAN
```

```python
heart=pd.read_csv('heart.csv')
```

```python
XDB=heart.iloc[:,:].values
```

```python
XDB.shape
```

```python
YDB = DBSCAN(eps=0.5, min_samples=5)
y_predDB = YDB.fit_predict(XDB)
```

```python
np.unique(y_predDB)
```

```python
plt.scatter(XDB[y_predDB == -1, 0], XDB[y_predDB == -1, 1], c='black',  label='Heart Disease')

plt.xlabel('Age')
plt.ylabel('Cholestral')
plt.title('DBSCAN CLUSTERING FOR HEART DISEASE DATASET')
plt.legend()

plt.show()
```
